```python
#Data cleaning code
def check_university(account_name):
    if isinstance(account_name, str):
        account_name_lower = account_name.lower()
        return 'university' in account_name_lower or 'univ' in account_name_lower or 'uni' in account_name_lower
    else:
        return False

# Apply the function to set the 'type of institution' value
df['Type of Institution'] = df.apply(lambda row: 'Academia' if check_university(row['Institution Name']) and pd.isnull(row['Type of Institution']) else row['Type of Institution'], axis=1)
df
```

```python
#Also data cleaning code
def is_gov(email_address):
    if isinstance(email_address, str):
        return email_address.upper().endswith("MIL") or email_address.upper().endswith("GOV")
    else:
        return False

# Apply the function to set the 'type of institution' value
df['Type of Institution'] = df.apply(lambda row: 'Government' if is_gov(row['Email']) and pd.isnull(row['Type of Institution']) else row['Type of Institution'], axis=1)
df

newdf = df['Type of Institution'].value_counts()
newdf
```

```python
#CODE FOR IDENTIFICATION CODE


df["Identifier Code"] = ''
def sorter(insti_type):
#python doesnt have switch case so we use this
    switcher = {
        "ACADEMIA": "ACA",
        "GOVERNMENT": "GOV",
        "CRO": "CRO",
        "HOSPITAL": "HOS",
        "PHARMACEUTICAL": "PHA",
        "PRIVATE RESEARCH ORGANIZATION": "PRO",
        "OTHER": "OTH",
        "VENDOR": "VEN",
        "VET PRACTICE": "VET",
        "MEDICAL DEVICE": "DEV"
    }
    if isinstance(insti_type,str):
      return switcher.get(insti_type.upper(), "XXX")
    else:
      return "XXX"

#insti code
df['Identifier Code'] = df['Type of Institution'].apply(sorter)

df['Identifier Code'] = df["Identifier Code"] + "/"
df


def ifUS(country):

    switcher = {
        "UNITED STATES": "U",
    }
    if isinstance(country,str):
      return switcher.get(country.upper(), "Z")
    else:
      return "Z"


# usa check code
df['Identifier Code'] = df ['Identifier Code'] + df['Country'].apply(ifUS)

df['Identifier Code'] = df["Identifier Code"] + "/"
df

#Put an X in the column desginated for specific animal if it is not found in the species field
def animal_code(animals):
    ani = ""
    if isinstance(animals,str):
      if ('RAT' in animals.upper()) : ani += 'R'
      else: ani += 'X'
      if ('MICE' in animals.upper()) : ani += 'M'
      else: ani += 'X'
      if ('FERRET' in animals.upper()) : ani += 'F'
      else: ani += 'X'
      if ('PRIMATE' in animals.upper()) : ani += 'P'
      else: ani += 'X'

      return ani
    else:
      return "XXXX"

# animal code
df['Identifier Code'] = df ['Identifier Code'] + df['Laboratory Species'].apply(animal_code)

#CODE FOR TIMEZONE
#creating timezone field

df['Time Zone'] = ''




state_dict = {
    'NY': 'EST',
    'MA': 'EST',
    'PA': 'EST',
    'MD': 'EST',
    'NC': 'EST',
    'SC': 'EST',
    'CA': 'PST',
    'AL': 'CST',
    'AZ': 'MST',
    'OH': 'EST',
    'TX': 'CST',
    'SC': 'EST',
    'OK': 'CST',
    'IA': 'CST',
    'MI': 'EST',
    'NE': 'CST',
    'UT': 'MST',
    'WA': 'PST',
    'LA': 'CST',
    'IL': 'CST',
    'MN': 'CST',
    'CT': 'EST',
    'FL': 'EST',
    'KS': 'CST',
    'MO': 'CST',
    'GA': 'EST',
    'WI': 'CST',
    'NJ': 'EST',
    'MS': 'CST',
    'CO': 'MST',
    'TN': 'CST',
    'VA': 'EST',
    'IN': 'EST',
    'NM': 'MST',
    'OR': 'PST',
    'RI': 'EST',
    'NV': 'PST',
    'KY': 'EST',
    'VT': 'EST',
    'MT': 'MST',
    'DE': 'EST',
    'AR': 'CST',
    'HI': 'HST',
    'ID': 'MST',
    'WV': 'EST',
    'ME': 'EST',
    'NH': 'EST',
    'SD': 'MST',
    'AK': 'AKST',
    'DC': 'EST',
    #'AB': 'EST',
    'WY': 'EST',
    #'ON': 'EST',
    'ND': 'CST',
    #'AE': 'EST',
    #'AP': 'EST',


}
def assign_timeZone(row):
  if row['Country'] == "United States":
    tz =  state_dict.get(row['State'], 'UNK') #if in the USA but dont know state, we put UNK
    return tz
  else:
    tz = "XXX" # if not in the USA put XXX

    return tz

#filling in the timezone field
df['Time Zone'] = df.apply(assign_timeZone, axis = 1)

#updating ID code with timezone as well
df['Identifier Code'] = df["Identifier Code"] + "/"
df['Identifier Code'] = df['Identifier Code'] + df['Time Zone']
df
```
